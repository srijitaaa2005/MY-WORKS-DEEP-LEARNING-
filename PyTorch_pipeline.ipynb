```python
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace = True)
df.head()
```

```python
df.shape
```

```python
df['diagnosis'].value_counts()
```

```python
y=df['diagnosis']
X=df.iloc[:,1:]

X
```

```python
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
X_train
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
y_test
```

```python
if torch.cuda.is_available():
  print("GPU is available")
  print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
  print("GPU is not available")
```

**NUMPY ARRAYS TO PYTORCH TENSORS**

```python
X_train=torch.from_numpy(X_train).float()
X_test=torch.from_numpy(X_test).float()
y_train=torch.from_numpy(y_train).float()
y_test=torch.from_numpy(y_test).float()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**DATASET & DATALOADER**

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
  def __init__(self,features,labels):
    self.features=features
    self.labels=labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
train_dataset=CustomDataset(X_train, y_train)
test_dataset=CustomDataset(X_test, y_test)
```

```python
test_dataset[10]
```

**CREATE DATALOADER**

```python
train_loader=DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader=DataLoader(test_dataset, batch_size=32, shuffle=True)
```

**DEFINNING THE MODEL**

```python
import torch.nn as nn
```

**IMPORTANT PARAMETERS**

```python
learning_rate=0.1
epochs=30
```

```python
loss_function=nn.BCELoss()
```

**TRAINING PIPELINE**

```python
# #converting training and testing data to float 32 from 64 since nn.Linear layer weights are float32 (Float) by default
# X_train = X_train.float()
# y_train = y_train.float()
# X_test = X_test.float()
# y_test = y_test.float()

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

#create model and move to GPU
model = Model(X_train.shape[1]).to(device)

#move data to GPU
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)

#create optimizer
optimizer=torch.optim.SGD(model.parameters(), lr=learning_rate)

for i in range(epochs):

  for batch_features, batch_labels in train_loader:
    batch_features = batch_features.to(device)
    batch_labels = batch_labels.to(device)

    #forward pass
    '''y_pred=model.forward_pass(X_train)'''
    '''y_pred=model(X_train)'''
    y_pred=model(batch_features)

    #loss calculate
    '''loss=model.loss_function(y_pred,y_train)'''
    '''loss=loss_function(y_pred,y_train.view(-1,1))'''
    loss=loss_function(y_pred,batch_labels.view(-1,1))

    #clear gradient
    optimizer.zero_grad()

    #backward pass
    loss.backward()

    #parameters update
    '''with torch.no_grad():
      model.linear.weight -= lr*model.linear.weight.grad
      model.linear.bias -= lr*model.linear.bias.grad'''

    optimizer.step()

    #zero gradient
    '''model.linear.weight.grad.zero_()
    model.linear.bias.grad.zero_()'''

  #print loss for each epoch
  if (i + 1) % 10 == 0:
    print(f'Epoch: {i+1} Loss: {loss.item()}')
```
